```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 1
b = 2
N = 10
x = np.linspace(0, a, N+1)
y = np.linspace(0, b, N+1)
print(x)
print(y)
```

```python
# https://www.sharpsightlabs.com/blog/numpy-meshgrid/
X, Y = np.meshgrid(x, y, sparse=False)
print(X)
print(Y)
```

```python
f = X**2 * Y**2
f
```

```python
plt.contour(X,Y,f) # you can add levels as attr
plt.xlabel("X")
plt.ylabel("Y")
plt.colorbar(label="f")
```

```python
plt.contourf(X,Y,f)
plt.xlabel("X")
plt.ylabel("Y")
plt.colorbar(label="f")
```

```python
fig = plt.figure()
 
# syntax for 3-D projection
ax = plt.axes(projection ='3d')
 
# plotting
ax.scatter(X, Y, f)
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
# linspace vs arange

a = np.linspace(0, 10, num=5)
# linspace generated `num` number of value between start and stop 
print(a)

b = np.arange(0, 10, step=2.5) # stop is excluded
# linspace generated a bunch of values between start and stop, 
# separated by a step value 
print(b)
```

```python

# How the decision boundary is plotted in KNN
# Notice the shape at every step

a = np.arange(0, 5, step=1)
b = np.arange(10, 16, step=1)
print(a)
print(a.shape)
print(b)
print(b.shape)
```

```python
aa, bb = np.meshgrid(a, b)
print(aa)
print(aa.shape)
print(bb)
print(bb.shape)
Z = np.sin(np.sqrt(X**2 + Y**2))
print(Z.shape)
```

```python
print(aa.ravel())
print(aa.ravel().shape)
print(bb.ravel())
print(bb.ravel().shape)
```

```python
print(np.c_[aa.ravel(), bb.ravel()])
print(np.c_[aa.ravel(), bb.ravel()].shape)
```

```python
import random
Z = np.array([random.randint(0,1) for i in np.c_[aa.ravel(), bb.ravel()]])
print(Z)
print(Z.shape)
```

```python
Z = Z.reshape(aa.shape)
print(Z)
print(Z.shape)

# Now with aa, bb and Z, the contour is plotted
```
